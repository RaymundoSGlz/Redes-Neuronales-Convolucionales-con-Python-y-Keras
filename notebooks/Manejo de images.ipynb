```python
# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

## Cargamos las imágenes

```python
im = io.imread('../data/mara.jpeg')
im
```

```python
# Mostremos la imagen
plt.imshow(im)
plt.axis('off')  # Desactivamos los ejes
plt.show()
```

## Separando canales de color

```python
# Separemos los canales de la imagen
r = im[:, :, 0]
g = im[:, :, 1]
b = im[:, :, 2]
```

```python
# Mostramos la imagen en escala de grises de un canal con una rotación de 90 grados
plt.imshow(r.T, cmap='gray'); # cmap='gray' es para que se muestre en escala de grises
```

## RGB con un solo canal activo

```python
im.shape
```

```python
# Creamos una matriz de ceros del mismo tamaño de la imagen
im_aux = np.zeros(r.shape, dtype=np.uint8)

# Utilizamos la matriz de ceros para crear una imagen roja
red = np.dstack((r, im_aux, im_aux)).astype(np.uint8)
# Utilizamos la matriz de ceros para crear una imagen verde
green = np.dstack((im_aux, g, im_aux)).astype(np.uint8)
# Utilizamos la matriz de ceros para crear una imagen azul
blue = np.dstack((im_aux, im_aux, b)).astype(np.uint8)

# Concatenamos las imágenes
im_rgb = np.hstack((red, green, blue))

# Mostramos la imagen
plt.imshow(im_rgb)
plt.axis('off');  # Desactivamos los ejes
```
