## Importando librer√≠as

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
from skimage import io, color
```

## Carga de imagen

```python
img = io.imread('../data/yo.png')
img_gray = color.rgb2gray(img) # Convertir a escala de grises
```

```python
print(img_gray.shape)
plt.imshow(img_gray);
```

## Creaci√≥n de kernel y aplicaci√≥n

```python
# Definimos una funci√≥n para aplicar un kernel a una imagen
def apply_kernel(img: np.ndarray, kernel: np.ndarray):
    """Aplica un kernel a una imagen
    :param img: Imagen a la que se le aplicar√° el kernel
    :param kernel: Kernel a aplicar"""
    img_out = ndi.convolve(img_gray, kernel) # Aplicar el kernel a la imagen
    fig, axes = plt.subplots(1, 2, figsize=(10, 5)) # Crear figura
    axes[0].imshow(img_gray, cmap='gray') # Mostrar imagen original
    axes[0].set_title('Imagen original')
    axes[0].axis('off')

    axes[1].imshow(img_out, cmap='gray') # Mostrar imagen con kernel aplicado
    axes[1].set_title('Imagen con kernel aplicado')
    axes[1].axis('off')
    plt.show()
```

```python
# Definimos un kernel de 3x3
kernel = np.array([[-1, 0, 1],
                   [-1, 0, 1],
                   [-1, 0, 1]])

# Aplicamos el kernel a la imagen
apply_kernel(img_gray, kernel)
```

```python
kernel = np.array([[1, 1, 1],
                   [0, 0, 0],
                   [-1, -1, -1]])

# Aplicamos el kernel a la imagen
apply_kernel(img_gray, kernel)
```
